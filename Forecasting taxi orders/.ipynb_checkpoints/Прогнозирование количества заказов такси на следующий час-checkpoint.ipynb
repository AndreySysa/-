```python
def plot_dataframe(df, column):
    # Создание графиков
    fig, axes = plt.subplots(1, 2, figsize=(15, 5))

    # Box plot для столбца
    axes[0].set_title(
        f'Boxplot для столбца {column}', fontsize=15, color='DarkSlateGray')
    sns.boxplot(data=df, y=column, ax=axes[0])
    axes[0].set_xlabel('Количество', fontsize=15, color='DarkSlateGray')
    axes[0].set_ylabel(column, fontsize=15, color='DarkSlateGray')
    axes[0].grid(True)
    axes[0].minorticks_on()
    axes[0].grid(which='minor', linestyle=':')

    # Histogram plot для столбца
    axes[1].set_title(
        f'Histogram plot для столбца {column}', fontsize=15, color='DarkSlateGray')
    sns.histplot(df[column], kde=False, bins=50, ax=axes[1])
    axes[1].set_ylabel('Количество', fontsize=15, color='DarkSlateGray')
    axes[1].set_xlabel(column, fontsize=15, color='DarkSlateGray')

    # Отображение графиков
    plt.minorticks_on()
    plt.grid(which='minor', linestyle=':')
    plt.grid(True)
    plt.show()

    # Подсчет уникальных значений и их количества
    unique_vals = df[column].unique()
    count = df[column].value_counts()
    describ = df[column].describe()
    print(describ)
    print('*'*100)
    print(f"Уникальные значения столбца {column}: {unique_vals}")
    print('*'*100)
    print(f"Количество каждого уникального значения: \n{count}")
```

```python
def plot_seasonal_decomposition(data, start_date, end_date):
    decomposed = seasonal_decompose(data[start_date:end_date])
    fig = plt.figure(figsize=(15, 15))
    plt.subplot(311)
    decomposed.trend.plot(ax=plt.gca(), color='green')
    plt.title('Тренд с {} по {}'.format(start_date, end_date),
              fontsize=20, color='DarkSlateGray')
    plt.ylabel('Кол-во заказов', fontsize=15, color='DarkSlateGray')
    plt.tight_layout()
    plt.minorticks_on()
    plt.grid(which='minor', linestyle=':')
    plt.grid(True)
    plt.subplot(312)
    decomposed.seasonal.plot(ax=plt.gca(), color='orange')
    plt.title('Сезонная составющая с {} по {}'.format(
        start_date, end_date), fontsize=20, color='DarkSlateGray')
    plt.ylabel('Кол-во заказов', fontsize=15, color='DarkSlateGray')
    plt.tight_layout()
    plt.minorticks_on()
    plt.grid(which='minor', linestyle=':')
    plt.grid(True)
    plt.subplot(313)
    decomposed.resid.plot(ax=plt.gca(), color='red')
    plt.title('Остаток декомпозиции с {} по {}'.format(
        start_date, end_date), fontsize=20, color='DarkSlateGray')
    plt.ylabel('Кол-во заказов', fontsize=15, color='DarkSlateGray')
    plt.tight_layout()
    plt.minorticks_on()
    plt.grid(which='minor', linestyle=':')
    plt.grid(True)
    plt.show()
```

Добавим новые признаки:
- год. не подходит. У нас данных меньше года.
- месяц. аналогично году. сезонность времени года будет просто в тренде уже заложена.
- день. не подходит.
- день недели. подходит, т.к. у нас есть недельная сезонность.
- час. подходит, т.к. есть часовая (суточная) сезонность.
- lag. подходит. это параметр авторегрессии - учёт влияния вчерашних событий на сегодняшние.
- rolling. подходит. это аналог тренда. Чем больше окно сглаживания, тем ровнее линия тренда.


## Обучите разные модели с различными гиперпараметрами.

### Создадим функцию для создания признаков.

```python
def make_features(data, max_lag, rolling_mean_size):
    tax = data
    tax['dayofweek'] = tax.index.dayofweek
    tax['hour'] = tax.index.hour

    for lag in range(1, max_lag + 1):
        tax['lag_{}'.format(lag)] = tax['num_orders'].shift(lag)

    tax['rolling_mean'] = tax['num_orders'].shift().rolling(
        rolling_mean_size).mean()
    return tax.dropna()
```

```python
%%time

def objective(trial):
    max_lag = trial.suggest_int('max_lag', 1, n_samples - 1)
    rolling_mean_size = trial.suggest_int('rolling_mean_size', 1, n_samples - 1)
    n_estimators = trial.suggest_int('n_estimators', 10, 100)
    max_depth = trial.suggest_int('max_depth', 1, 30)
    min_samples_split = trial.suggest_int('min_samples_split', 2, 10)
    min_samples_leaf = trial.suggest_int('min_samples_leaf', 1, 10)

    df = make_features(taxi, max_lag, rolling_mean_size)
    train, test = train_test_split(df, test_size=0.2, shuffle=False)
    valid, test = train_test_split(df, test_size=0.5, shuffle=False)

    features_train = train.drop(['num_orders'], axis=1)
    target_train = train['num_orders']

    features_valid = valid.drop(['num_orders'], axis=1)
    target_valid = valid['num_orders']

    model = RandomForestRegressor(n_estimators=n_estimators, max_depth=max_depth,
                                   min_samples_split=min_samples_split, min_samples_leaf=min_samples_leaf,
                                   n_jobs=-1, random_state=RANDOM_STATE)
    model.fit(features_train, target_train)
    predictions = model.predict(features_valid)

    rmse = mean_squared_error(target_valid, predictions, squared=False)

    return rmse

n_samples = int(len(taxi)*0.01)
study = optuna.create_study(direction="minimize")

study.optimize(objective, n_trials=100, n_jobs=-1)

best_max_lag = study.best_params['max_lag']
best_rolling_mean_size = study.best_params['rolling_mean_size']
best_n_estimators = study.best_params['n_estimators']
best_max_depth = study.best_params['max_depth']
best_min_samples_split = study.best_params['min_samples_split']
best_min_samples_leaf = study.best_params['min_samples_leaf']
best_rmse = study.best_value

print('лучшее RMSE =', best_rmse)
print('best_max_lag =', best_max_lag)
print('best_rolling_mean_size =', best_rolling_mean_size)
print('best_n_estimators =', best_n_estimators)
print('best_max_depth =', best_max_depth)
print('best_min_samples_split =', best_min_samples_split)
print('best_min_samples_leaf =', best_min_samples_leaf)
Forest = RandomForestRegressor(n_estimators=best_n_estimators,
                               max_depth=best_max_depth,
                               min_samples_split=best_min_samples_split,
                               min_samples_leaf=best_min_samples_leaf,
                               random_state=RANDOM_STATE,
                               n_jobs=-1)
```

### LGBMRegressor.

```python
%%time

def objective(trial):
    max_lag = trial.suggest_int('max_lag', 1, n_samples - 1)
    rolling_mean_size = trial.suggest_int(
        'rolling_mean_size', 1, n_samples - 1)
    n_estimators = trial.suggest_int('n_estimators', 10, 100)
    max_depth = trial.suggest_int('max_depth', 1, 30)
    learning_rate = trial.suggest_loguniform('learning_rate', 0.001, 0.1)

    df = make_features(taxi, max_lag, rolling_mean_size)
    train, test = train_test_split(df, test_size=0.2, shuffle=False)
    valid, test = train_test_split(df, test_size=0.5, shuffle=False)

    features_train = train.drop(['num_orders'], axis=1)
    target_train = train['num_orders']

    features_valid = valid.drop(['num_orders'], axis=1)
    target_valid = valid['num_orders']

    params = {
        "boosting_type": "gbdt",
        "objective": "regression",
        "metric": "rmse",
        "learning_rate": learning_rate,
        "n_estimators": n_estimators,
        "max_depth": max_depth,
        "num_leaves": trial.suggest_int("num_leaves", 10, 200),
        "min_child_samples": trial.suggest_int("min_child_samples", 5, 100),
        "subsample": trial.suggest_float("subsample", 0.1, 1.0),
        "subsample_freq": trial.suggest_int("subsample_freq", 1, 10),
        "feature_fraction": trial.suggest_float("feature_fraction", 0.1, 1.0),
        "lambda_l1": trial.suggest_float("lambda_l1", 0, 1),
        "lambda_l2": trial.suggest_float("lambda_l2", 0, 1),
        "random_state": RANDOM_STATE,
        "n_jobs": -1,
    }

    model = lgb.LGBMRegressor(**params)
    model.fit(features_train, target_train,
              eval_set=[(features_valid, target_valid)],
              early_stopping_rounds=10, verbose=False)

    predictions = model.predict(features_valid)

    rmse = mean_squared_error(target_valid, predictions, squared=False)

    return rmse


n_samples = int(len(taxi)*0.01)
study = optuna.create_study(direction="minimize")

study.optimize(objective, n_trials=100, n_jobs=-1)

best_max_lag = study.best_params['max_lag']
best_rolling_mean_size = study.best_params['rolling_mean_size']
best_n_estimators = study.best_params['n_estimators']
best_max_depth = study.best_params['max_depth']
best_learning_rate = study.best_params['learning_rate']
best_rmse = study.best_value

print('лучшее RMSE =', best_rmse)
print('best_max_lag =', best_max_lag)
print('best_rolling_mean_size =', best_rolling_mean_size)
print('best_n_estimators =', best_n_estimators)
print('best_max_depth =', best_max_depth)
print('best_learning_rate =', best_learning_rate)

model_params = {
    "boosting_type": "gbdt",
    "objective": "regression",
    "metric": "rmse",
    "learning_rate": best_learning_rate,
    "n_estimators": best_n_estimators,
    "max_depth": best_max_depth,
    "num_leaves": study.best_params["num_leaves"],
    "min_child_samples": study.best_params["min_child_samples"],
    "subsample": study.best_params["subsample"],
    "subsample_freq": study.best_params["subsample_freq"],
    "feature_fraction": study.best_params["feature_fraction"],
    "lambda_l1": study.best_params["lambda_l1"],
    "lambda_l2": study.best_params["lambda_l2"],
    "random_state": RANDOM_STATE,
    "n_jobs": -1,
}

LGBM = lgb.LGBMRegressor(**model_params)
```

### CatBoostRegressor.

```python
%%time

def objective(trial):
    max_lag = trial.suggest_int('max_lag', 1, n_samples - 1)
    rolling_mean_size = trial.suggest_int(
        'rolling_mean_size', 1, n_samples - 1)
    learning_rate = trial.suggest_loguniform('learning_rate', 0.001, 0.1)
    n_estimators = trial.suggest_int('n_estimators', 10, 100)
    max_depth = trial.suggest_int('max_depth', 1, 8)

    df = make_features(taxi, max_lag, rolling_mean_size)
    train, test = train_test_split(df, test_size=0.2, shuffle=False)
    valid, test_ = train_test_split(test, test_size=0.5, shuffle=False)

    X_train = train.drop(['num_orders'], axis=1)
    y_train = train['num_orders']

    X_valid = valid.drop(['num_orders'], axis=1)
    y_valid = valid['num_orders']

    model_params = {
        'iterations': n_estimators,
        'learning_rate': learning_rate,
        'depth': max_depth,
        'l2_leaf_reg': trial.suggest_float('l2_leaf_reg', 0.1, 10.0),
        'border_count': trial.suggest_int('border_count', 5, 255),
        'bagging_temperature': trial.suggest_float('bagging_temperature', 0, 1),
        'random_strength': trial.suggest_float('random_strength', 0, 1),
        'early_stopping_rounds': 10,
        'random_seed': RANDOM_STATE,
        'thread_count': -1,
        'verbose': False
    }

    model = CatBoostRegressor(**model_params)
    model.fit(X_train, y_train, eval_set=[(X_valid, y_valid)])

    predictions = model.predict(X_valid)
    rmse = mean_squared_error(y_valid, predictions, squared=False)

    return rmse


n_samples = int(len(taxi)*0.01)
study = optuna.create_study(direction="minimize")

study.optimize(objective, n_trials=100, n_jobs=-1)

best_max_lag = study.best_params['max_lag']
best_rolling_mean_size = study.best_params['rolling_mean_size']
best_n_estimators = study.best_params['n_estimators']
best_max_depth = study.best_params['max_depth']
best_learning_rate = study.best_params['learning_rate']
best_rmse = study.best_value

print('лучшее RMSE =', best_rmse)
print('best_max_lag =', best_max_lag)
print('best_rolling_mean_size =', best_rolling_mean_size)
print('best_n_estimators =', best_n_estimators)
print('best_max_depth =', best_max_depth)
print('best_learning_rate =', best_learning_rate)

model_params = {
    'iterations': best_n_estimators,
    'learning_rate': best_learning_rate,
    'depth': best_max_depth,
    'l2_leaf_reg': study.best_params['l2_leaf_reg'],
    'border_count': study.best_params['border_count'],
    'bagging_temperature': study.best_params['bagging_temperature'],
    'random_strength': study.best_params['random_strength'],
    'early_stopping_rounds': 10,
    'random_seed': RANDOM_STATE,
    'thread_count': -1,
    'verbose': False
}

CatBoost = CatBoostRegressor(**model_params)
```

```python
df = make_features(taxi, best_max_lag, best_rolling_mean_size)
train, test = train_test_split(df, test_size=0.2, shuffle=False)
valid, test = train_test_split(test, test_size=0.5, shuffle=False)

features_train = train.drop(['num_orders'], axis=1)
target_train = train['num_orders']

features_valid = valid.drop(['num_orders'], axis=1)
target_valid = valid['num_orders']

model = CatBoost
model.fit(features_train, target_train)
predictions = model.predict(features_valid)

rmse = mean_squared_error(target_valid, predictions, squared=False)
# визуализирую реальные и предсказанные значения тестовой выборки:
fig, ax = plt.subplots(figsize=(15, 5))
ax.plot(target_valid.index, target_valid, label='реальное кол-во заказов')
ax.plot(target_valid.index, predictions, label='предсказанное кол-во заказов')
plt.title('Реальные и предсказанные CatBoostRegressor значения валидационной выборки',
          fontsize=20, color='DarkSlateGray')
plt.xlabel('Даты', fontsize=15, color='DarkSlateGray')
plt.ylabel('Кол-во заказов', fontsize=15, color='DarkSlateGray')
plt.legend()
plt.tight_layout()
plt.minorticks_on()
plt.grid(which='minor', linestyle=':')
plt.grid(True)
plt.show()
```
